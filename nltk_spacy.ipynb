```python
text = "NLP is rapidly evolving with research aimed at refining models to comprehend things like sarcasm and cultural references."
text
```

```python
# Download a pre-trained punkt tokenizer for English
# To execute these commands, root certificates are required
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# Import a function word_tokenize from nltk.tokenize module
from nltk.tokenize import word_tokenize

# Break the sentence into words
tokens = word_tokenize (text)
print(tokens)
# tokens  # to display in multiple lines
```

```python
# Display stem for each word
# PorterStemmer is an algorithm used for stemming in English
from  nltk.stem import PorterStemmer
stemmer = PorterStemmer()

# Apply stemming on each word in the token list
stemmed_words = [stemmer.stem(token) for token in tokens]
print(stemmed_words)

# stemmed_words # to display in multiple lines
```

```python

# Part-of-speech tagging to determine the verbs, nouns, etc.,
from nltk import pos_tag

# pos_tag internally loads Perception POS tagging model
# so install the dataset
nltk.download('averaged_perceptron_tagger_eng')

text_pos_tag = pos_tag(tokens)
print(text_pos_tag)
text_pos_tag
```

```python
# Install resource tageset_json
nltk.download('tagsets_json')

# Display the label for each tag 
nltk.help.upenn_tagset("NNP")
```

```python
# Named Entity Recognition (NER)
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')

# Differentiate Dave(Person) from Paris(location)
entities = nltk.ne_chunk(text_pos_tag)
print(entities) # now Dave & Paris are recognised
```

```python
# SpaCy 3.x officially supports Python 3.7вЂ“3.11.
!pip3 install --upgrade pip
!pip3 install "spacy<4.0" --force-reinstall

import spacy
# Ensure you have this model downloaded
!python -m spacy download en_core_web_lg

# Load the English Language pipeline for optimizing CPU performance
# Pre-trained word embedding
nlp = spacy.load('en_core_web_lg')

# Word embeddings are a type of word representation where words 
# are mapped to vectors allowing words with similar meaning to 
# have a similar representation

# Process words with the model
word1 = nlp("king")
word2 = nlp("queen")
word3 = nlp("apple")

# Calculate the similarities(or word embeddings)
similarity1 = word1.similarity(word2)
similarity2 = word1.similarity(word3)

# Display the similarities
print(f"Similarity between 'king' and 'queen': {similarity1:.2f}")
print(f"Similarity between 'king' and 'apple': {similarity2:.2f}")

# King & Queen woudl show higher similarity due to its contextual closeness
```
